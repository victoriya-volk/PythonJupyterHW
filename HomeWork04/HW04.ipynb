## Условие 3: 3 задача

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

```python
import numpy as np
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
import seaborn as sns 

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
sns.jointplot(x = df['bathrooms'], y = df['price'], kind='reg')
```

#### Вывод по графику взаимосвязи количества ванных комнат и цены дома
Цена дома зависит от количества ванных комнат и это видно на графике: максимальная цена у домов с максимальным количеством комнат.

```python
sns.jointplot(x = df['floors'], y = df['price'], kind='reg')
```

#### Вывод по графику взаимосвязи цены и этажности дома
Прослеживается явная зависимость, при этом можно выделить максимальные цены у домов с 2-2.5 этажа.
